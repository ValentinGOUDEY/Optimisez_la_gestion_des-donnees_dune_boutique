### outliers

```python
#On affiche un nuage de point des prix des articles pour repéré d'eventuelle prix abérrant
df.plot.scatter(x='product_id', y='price', c = 'price', s=30, cmap = 'viridis');
```

On remarque au moins 8 prix qui sortent du lot

```python
#On affiche un boxplot des prix

sns.boxplot(x = df.price)
plt.title('Boxplot du Prix des bouteilles de vin')

#On definit une fonction qui calcul les prix qui sont en dehors de la distance interquartile, donc outliers
def outlierHandling(price):
    price = sorted(price)
    Q1 , Q3 = np.percentile(price, [25,75])
    print('Le premier quartile est de:',Q1,'\n''Le troisième quartile est de:',Q3)
    IQR = Q3 - Q1
    upperBound = Q3 + (1.5 * IQR)
    print('upperBound : ', upperBound)
    return upperBound

upperbound = outlierHandling(df.price)
outliers= df[(df.price > upperbound)]
```

Ce boxplot nous montre beaucoup de prix qui sont considéré comme outliers

```python
#On crée les variables quartiles Q1 et Q3
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

#On crée la variable distance interquartile (Inter-quartile range)
IQR = Q3 - Q1

#On crée une colonne 'outliers' qui definit par true ou false les outliers
df['outliers'] = (df.price > (Q3 + 1.5 * IQR))

#On affiche une visualisation des outliers
plt.scatter(df['product_id'],df['price'],s = 20,c = df['outliers'],cmap = 'RdYlGn')
plt.xlabel('product_id')
plt.ylabel('price')
plt.title('Visualisation des outliers')
plt.show() 
```

```python
#On affiche le nombre de prix considéré comme outliers
df.outliers.sum()
```

```python
#On affiche les 32 prix considérés comme outliers
outliers.head(32)
```

J'ai pris les produits qui ont un prix considéré comme outliers et j'ai comparé leurs prix sur internet et leurs prix sont cohérent, ce sont donc juste des vins qui sont plus chères dû à leurs qualité/rareté

```python
#on stock dans q les valeurs qui rentre dans le quantile 0.99 pour repérer les outliers les plus élevés
q = df["price"].quantile(0.99)
```

```python
#on stock dans outliers_price les valeurs plus élevé de la colonne 'price'
outliers_price = df[df["price"] > q]
```

```python
#On affiche les valeurs les plus élevé de la colonne 'price' 
outliers_price.sort_values(by = 'price', ascending = False)
```

Il y a bien 8 produits qui en ressort

```python
#On affiche la liste des produits qui ont un prix les plus élevé
sns.barplot(y=outliers_price.price, x=outliers_price.product_id, order = outliers_price.sort_values('price', ascending=False).product_id)
plt.title('Liste des produits avec un prix les plus élevé')
plt.show()
```
